```python
import pandas as pd

# 1) Pickle 파일 6개 불러오기
df1 = pd.read_pickle("preprocessing/demographic/df_demo.pkl")
df2 = pd.read_pickle("preprocessing/examination/df_exam.pkl")
df3 = pd.read_pickle("preprocessing/lab/df_lab.pkl")
df4 = pd.read_pickle("preprocessing/label/df_true.pkl")
df5 = pd.read_pickle("preprocessing/question/df_ques.pkl")

# 2) 리스트로 묶기
dfs = [df1, df2, df3, df4, df5]

# 3) 공통 병합 key (예: SEQN)
key = "subject_no"   # 네 데이터에서 공통으로 존재하는 컬럼명으로 수정하면 됨

# 4) reduce 형태로 순차 병합
from functools import reduce

df_merged = reduce(lambda left, right: pd.merge(left, right, on=key, how='inner'), dfs)

# 5) 결과 확인
print(df_merged.shape)
df_merged.head()
```
