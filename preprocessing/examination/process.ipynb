```python
#data load, concat(ref=SEQN), dropna

import pandas as pd
from functools import reduce

#data load, feature name alteration
df1 = pd.read_sas(r"C:\Users\cmc\Desktop\monash AI\data\BMX_L.xpt", format='xport')
df_fil1=df1[['SEQN','BMXWT','BMXHT','BMXBMI','BMXWAIST','BMXHIP']]

df2 = pd.read_sas(r"C:\Users\cmc\Desktop\monash AI\data\BPXO_L.xpt", format='xport')
df_fil2=df2[['SEQN','BPXODI3','BPXOPLS3','BPXOSY3']] 


#data concat
dfs = [df_fil1, df_fil2]
df_merged = reduce(
    lambda left, right: pd.merge(left, right, on="SEQN", how="inner"),
    dfs
)

#data dropna
df_merged=df_merged.dropna()

#column rename
df_merged2 = df_merged.rename(columns={
    'SEQN': 'subject_no',
    'BMXWT': 'wt',
    'BMXHT': 'ht',
    'BMXBMI':'bmi',
    'BMXWAIST':'wst',
    'BMXHIP':'hip',
    'BPXODI3':'dia',
    'BPXOPLS3':'pulse',
    'BPXOSY3':'sys',
})

df_merged2.shape
```

```python
df_merged2
```

```python
#outlier extract
df_merged2 = df_merged2[df_merged2['wt'] <= 200.0]
df_merged2[['wt']].value_counts()  
counts = df_merged2['wt'].value_counts()
counts_sorted = counts.sort_index()
#print(counts_sorted)

df_merged2[['ht']].value_counts()  
counts = df_merged2['ht'].value_counts()
counts_sorted = counts.sort_index()
#print(counts_sorted)

df_merged2 = df_merged2[df_merged2['bmi'] <= 60.0]
df_merged2[['bmi']].value_counts()  
counts = df_merged2['bmi'].value_counts()
counts_sorted = counts.sort_index()
#print(counts_sorted)

df_merged2 = df_merged2[df_merged2['wst'] <= 160.0]
df_merged2[['wst']].value_counts()  
counts = df_merged2['wst'].value_counts()
counts_sorted = counts.sort_index()
#print(counts_sorted)

df_merged2 = df_merged2[df_merged2['hip'] <= 160.0]
df_merged2[['hip']].value_counts()  
counts = df_merged2['hip'].value_counts()
counts_sorted = counts.sort_index()
#print(counts_sorted)

df_merged2[['sys']].value_counts()  
counts = df_merged2['sys'].value_counts()
counts_sorted = counts.sort_index()
#print(counts_sorted)   #lower bound?

df_merged2[['dia']].value_counts()  
counts = df_merged2['dia'].value_counts()
counts_sorted = counts.sort_index()
#print(counts_sorted)

df_merged2 = df_merged2[df_merged2['pulse'] <= 130.0]
df_merged2[['pulse']].value_counts()  
counts = df_merged2['pulse'].value_counts()
counts_sorted = counts.sort_index()
print(counts_sorted)
```
