```python
#data load, concat(ref=SEQN), dropna

import pandas as pd
from functools import reduce

#data load, feature name alteration
df = pd.read_sas(r"C:\Users\cmc\Desktop\monash AI\data\DEMO_L.xpt", format='xport')
df_fil=df[['SEQN','RIAGENDR','RIDAGEYR','RIDRETH3','DMDEDUC2','DMDMARTZ','DMDHHSIZ','INDFMPIR']] 

df_fil=df_fil.dropna()

df_fil = df_fil.rename(columns={
    'SEQN': 'subject_no',
    'RIAGENDR': 'gender', 
    'RIDAGEYR': 'age',
    'RIDRETH3': 'race',
    'DMDEDUC2': 'educ',
    'DMDMARTZ': 'marry',
    'DMDHHSIZ': 'house',
    'INDFMPIR': 'pov',
})

df_fil.shape
```

```python
df_fil
```

```python
#outlier extract
#df_fil = df_fil[df_fil['alt'] <= 200.0]
df_fil[['gender']].value_counts()  
counts = df_fil['gender'].value_counts()
counts_sorted = counts.sort_index()
#print(counts_sorted)

df_fil[['age']].value_counts()  
counts = df_fil['age'].value_counts()
counts_sorted = counts.sort_index()
#print(counts_sorted)              #age: 20~80

df_fil[['race']].value_counts()  
counts = df_fil['race'].value_counts()
counts_sorted = counts.sort_index()
#print(counts_sorted)

df_fil = df_fil[df_fil['educ'] != 9.0]
df_fil[['educ']].value_counts()  
counts = df_fil['educ'].value_counts()
counts_sorted = counts.sort_index()
#print(counts_sorted)


df_fil = df_fil[df_fil['marry'] != 77.0]
df_fil = df_fil[df_fil['marry'] != 99.0]
df_fil[['marry']].value_counts()  
counts = df_fil['marry'].value_counts()
counts_sorted = counts.sort_index()
#print(counts_sorted)

df_fil[['house']].value_counts()  
counts = df_fil['house'].value_counts()
counts_sorted = counts.sort_index()
#print(counts_sorted)

df_fil[['pov']] = df_fil[['pov']].astype(float)
df_fil[['pov']].value_counts()  
counts = df_fil['pov'].value_counts()
counts_sorted = counts.sort_index()
print(counts_sorted)
```
