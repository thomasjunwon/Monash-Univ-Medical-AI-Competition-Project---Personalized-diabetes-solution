```python
#data load, concat(ref=SEQN), dropna

import pandas as pd
from functools import reduce

#data load, feature name alteration
df = pd.read_sas(r"C:\Users\cmc\Desktop\monash AI\data\DIQ_L.xpt", format='xport')
df_fil=df[['SEQN','DIQ010']] 


df_fil=df_fil.dropna()
print(df_fil[['DIQ010']].value_counts())



df_fil = df_fil.rename(columns={
    'SEQN': 'subject_no',
    'DIQ010': 'dbs'
})

df_fil.shape
```

```python
df_fil = df_fil[df_fil['dbs'] != 9.0]
df_fil[['dbs']].value_counts()  
counts = df_fil['dbs'].value_counts()
counts_sorted = counts.sort_index()
print(counts_sorted)
```

```python
#outlier extract
df_fil = df_fil[df_fil['dbs'] != 9.0]

replace_dict = {1.0: 2, 2.0: 0,3.0:1}

# 값 치환
df_fil['dbs'] = df_fil['dbs'].replace(replace_dict)

df_fil[['dbs']].value_counts()  
counts = df_fil['dbs'].value_counts()
counts_sorted = counts.sort_index()
print(counts_sorted)
```
