```python
#data load, concat(ref=SEQN), dropna

import pandas as pd
from functools import reduce

#data load, feature name alteration
df1 = pd.read_sas(r"C:\Users\cmc\Desktop\monash AI\data\BIOPRO_L.xpt", format='xport')
df_fil1=df1[['SEQN','LBXSATSI','LBXSAL','LBXSASSI','LBXSCR','LBXSCH','LBXSTR','LBXSGTSI']] #'LBXGLU', 'LBXIN','LBXHSCRP'

df2 = pd.read_sas(r"C:\Users\cmc\Desktop\monash AI\data\CBC_L.xpt", format='xport')
df_fil2=df2[['SEQN','LBXWBCSI','LBXHGB','LBXHCT']] #'LBXGLU', 'LBXIN','LBXHSCRP'

df3 = pd.read_sas(r"C:\Users\cmc\Desktop\monash AI\data\TRIGLY_L.xpt", format='xport')                 #데이터 너무 적음
df_fil3=df3[['SEQN','LBDLDLM']] #'LBXGLU', 'LBXIN','LBXHSCRP'

df4 = pd.read_sas(r"C:\Users\cmc\Desktop\monash AI\data\HDL_L.xpt", format='xport')
df_fil4=df4[['SEQN','LBDHDD']] 

df5 = pd.read_sas(r"C:\Users\cmc\Desktop\monash AI\data\ALB_CR_L.xpt", format='xport')
df_fil5=df5[['SEQN','URDACT']] 

df6 = pd.read_sas(r"C:\Users\cmc\Desktop\monash AI\data\GLU_L.xpt", format='xport')          #데이터 너무 적음
df_fil6=df6[['SEQN','LBXGLU']] 

df7 = pd.read_sas(r"C:\Users\cmc\Desktop\monash AI\data\INS_L.xpt", format='xport')
df_fil7=df7[['SEQN','LBXIN']] 

df8 = pd.read_sas(r"C:\Users\cmc\Desktop\monash AI\data\HSCRP_L.xpt", format='xport')
df_fil8=df8[['SEQN','LBXHSCRP']] 

#data concat
dfs = [df_fil1, df_fil2, df_fil3,df_fil4,df_fil5,df_fil6,df_fil7,df_fil8]
df_merged = reduce(
    lambda left, right: pd.merge(left, right, on="SEQN", how="inner"),
    dfs
)

#data dropna
df_merged=df_merged.dropna()

#column rename
df_merged2 = df_merged.rename(columns={
    'SEQN': 'subject_no',
    'LBXSATSI': 'alt',
    'LBXSAL': 'albumin',
    'LBXSASSI':'ast',
    'LBXSCR':'crea',
    'LBXSCH':'chol',
    'LBXSTR':'tyg',
    'LBXSGTSI':'ggt',
    'LBXGLU':'glu',
    'LBXHCT':'hct',
    'LBXHGB':'hb',
    'LBXWBCSI':'wbc',
    'LBDLDLM':'ldl',
    'LBDHDD':'hdl',
    'URDACT':'acratio',
    'LBXIN':'insulin',
    'LBXHSCRP':'crp'

})

df_merged2.shape
```
